in order to load dataset into sql database, data types of each column in dataset must be fix. here we need to check data types and change it if needed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset to check data type
df = pd.read_csv('DS_Job_Dataset.csv')
df.info()
```

Here we have a lot of columns that have object type, meaning the data type is not fix yet. we need to change it

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#change date column into datetime data type
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
df = df.astype({'job_title_short':'string',
                'job_title':'string',
                'job_location':'string',
																'job_via':'string',
																'job_schedule_type':'string',
																'search_location':'string',
																'job_country':'string',
																'salary_rate':'string',
																'company_name':'string',
																'skills':'string'})
df.info()
```

```python
df.sample(20)
```

alright every column in the right format, so next step is corvert to csv then upload to sql database

```python
df.info()
```

```python
# Create function to check wether there is ';' is data because it will damage csv output
def check_semicolon_in_df(df):
    problematic_columns = {}
    for col in df.columns:
        if pd.api.types.is_string_dtype(df[col]):
            count = df[col].str.contains(';', na=False).sum()         
            if count > 0:
                problematic_columns[col] = count
    
    return problematic_columns
```

```python
check_semicolon_in_df(df)
```

need to subtitute ';' to ',' instead

```python
fix_column = ['job_title', 'company_name']
```

```python
for i in fix_column:
    df[i] = df[i].str.replace(';', ',')
```

```python
check_semicolon_in_df(df)
```

ok everything is clear. time to convert to csv

```python
df.to_csv('Ready_DS_Job_Dataset.csv', index=False, sep=';', encoding='utf-8')
```
